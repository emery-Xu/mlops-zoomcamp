```python
import pandas as pd
```

```python
df = pd.read_parquet("yellow_tripdata_2023-03.parquet")
```

```python
df.head()
```

```python
print(len(df.columns))
len(df)
```

```python
type(df['tpep_pickup_datetime'])
```

```python
df['duration'] = (
    df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
).dt.total_seconds() / 60

std_duration = df['duration'].std()
print(std_duration)
```

```python
df = df[(df.duration >= 1) & (df.duration <= 60)]
len(df)
```

```python
print(3316216/3403766)
```

```python
categories = ['PULocationID', 'DOLocationID']
df[categories] = df[categories].astype(str)
```

```python
train_dict = df[categories].to_dict(orient="records")
train_dict[:7]
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer()
X = dv.fit_transform(train_dict)
X.shape
```

```python
from sklearn.linear_model import LinearRegression

y = df['duration'].values
lr = LinearRegression()
lr.fit(X, y)
```

```python
y_pred = lr.predict(X)
y_pred[:7]
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y, y_pred))
rmse
```
